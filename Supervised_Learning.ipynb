```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_excel("/content/diabetes.xls")
```

```python
df
```

```python
df.columns
```

```python
x=df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]
```

```python
y=df[['Outcome']]
```

```python
x
```

```python
y
```

```python
x_train, x_test  , y_train ,y_test=train_test_split(x,y , test_size=0.2)
```

```python
model1=DecisionTreeClassifier()
model2=RandomForestClassifier()
model3=KNeighborsClassifier()
model4=GradientBoostingClassifier()
```

```python
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train,y_train)
```

```python
model3.fit(x_train,y_train)
```

```python
model4=model4.fit(x_train,y_train)
```

```python
model1=model1.predict(x_test)
```

```python
model2=model2.predict(x_test)
```

```python
model3=model3.predict(x_test)
```

```python
model4=model4.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, model1)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, model1))
print('Confusion Matrix:')
print(confusion_matrix(y_test, model1))
```

```python
accuracy = accuracy_score(y_test, model2)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, model2))
print('Confusion Matrix:')
print(confusion_matrix(y_test, model2))
```

```python
accuracy = accuracy_score(y_test, model3)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, model3))
print('Confusion Matrix:')
print(confusion_matrix(y_test, model3))
```

```python
accuracy = accuracy_score(y_test, model4)
print(f'Accuracy: {accuracy}')
print('Classification Report:')
print(classification_report(y_test, model4))
print('Confusion Matrix:')
print(confusion_matrix(y_test, model4))
```
