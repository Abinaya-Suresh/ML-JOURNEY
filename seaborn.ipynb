```python
import seaborn as sns
```

```python
sns.scatterplot(x="X", y="price", data=df,hue="type")
```

```python
sns.histplot(df["X"],kde=True , bins=15)#kde -lining ,its line distplot which is nothing but distribution
```

```python
sns.displot(df["price"],kde=True , bins=15)
```

```python
sns.barplot(x="type",y="price",data=df)
```

```python
sns.boxplot(x="type",y="price" , data=df ,hue="type",palette="YlGnBu") #palette-color
```

```python
sns.stripplot(x="type",y="price" , data=df ,hue="size" )
```

```python
sns.stripplot(x="type",y="price" , data=df ,hue="size" , dodge=True)#dodgr-is used to seperate the different types of relations
```

```python
#jointplot=scatterplot and distributionplot
sns.jointplot(x="X",y="price" , data=df )
```

```python
sns.jointplot(x="X",y="price" , data=df ,kind="reg")
```

```python
sns.jointplot(x="X",y="price" , data=df ,kind="hex")
```

```python
sns.pairplot(df.select_dtypes(['number']))
```

```python
sns.relplot(x="X",y="price",data=df)
```

```python
sns.heatmap(df.select_dtypes(['number']).corr(),cmap="icefire")
```
